```python
n = 22

def get_box():
    box = [0]*n
    for i_x in range(n):
        box[i_x] = [0]*n
    return box

def is_safe(box, row, col):
    for i in range(col):
        if box[row][i] == 1:
            return False
 
    for i,j in zip(range(row,-1,-1), range(col,-1,-1)):
        if box[i][j] == 1:
            return False
 
    for i,j in zip(range(row,n,1), range(col,-1,-1)):
        if box[i][j] == 1:
            return False
    return True


def solve(box, col):
    if col >= n:
        return True
    
    for i in range(n):
        if is_safe(box, i, col):
            box[i][col] = 1
            if col == n-1 :
                for x in range(n):
                    for y in range(n):
                        if box[x][y] == 1:
                            print('(',x, ',',y,')',end='')
                print()
                box[i][col] = 0
                return 
            
            solve(box, col+1)     
            box[i][col] = 0
            
```
